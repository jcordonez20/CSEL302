# **1. Import Libraries**

```python
import pandas as pd #Data Manipulation
import numpy as np #Numerical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Data Modeling
```

# **2. Generating Random Data**

```python
data = pd.read_csv('/content/Lesson_4_student_grades_dataset.csv', encoding='latin-1')
data.head()
```

# **3. Visualizing the Data**

```python
plt.scatter(data['Previous_Exam_Scores'],data['Grades'])
plt.xlabel('Previous Exam Scores')
plt.ylabel('Grades')
plt.show()
```

# **4. Preparing Data for Modeling**

```python
X = data[['Previous_Exam_Scores']] #Independent Variable
Y = data[['Grades']] #Dependent Variable
```

# **5. Fitting the Linear Regression Model**

```python
model = LinearRegression()

model.fit(X,Y)
```

# **6. Getting Model Coefficients**

```python
slope = model.coef_[0]
intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

# **7. Visualizing the Linear Regression Line**

```python
plt.scatter(X,Y)
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Previous Exam Scores')
plt.ylabel('Grades')
plt.show()
```

# **8. Making Predictions**

```python
new_X = [[np.random.randint(1,100)]]
predicted_Y = model.predict(new_X)
print('Previous Exam Score: ',new_X)
print('Predicted Grade: ',predicted_Y)
```
